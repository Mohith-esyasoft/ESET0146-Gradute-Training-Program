```python
x = [0,2,4,6,8]
y=[0,4,16,36,64]

fig, ax = plt.subplots()
ax.plot(x,y, marker='o', label = "datapoints")
ax.set_title("Basic Component of Matplotlib Figure")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.legend()
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns       
arr = np.array([10,20,30])
df = pd.DataFrame({"Numbers":arr})
plt.bar(df.index, df["Numbers"])
plt.show()
```

```python
#  create a dictionary, covert it to dataframe and plot a bar graph
dictionary = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
              'Age': [24, 27, 22, 32],
                'City': ['New York', 'Los Angeles', 'Chicago', 'Houston']}
df = pd.DataFrame(dictionary)
plt.bar(df['Name'], df['Age'])
plt.xlabel('Name')
plt.ylabel('Age')
plt.title('Age of Individuals')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('integer.csv')
plt.boxplot(data['AZ'])
plt.title('Box Plot of AZ')
plt.ylabel('AZ')
plt.show()

# Outlier detection for AZ
Q1 = data['AZ'].quantile(0.25)
Q3 = data['AZ'].quantile(0.75)
IQR = Q3 - Q1
outliers = data[(data['AZ'] < Q1 - 1.5 * IQR) | (data['AZ'] > Q3 + 1.5 * IQR)]
print("Outliers in AZ:", outliers['AZ'].values)
```
